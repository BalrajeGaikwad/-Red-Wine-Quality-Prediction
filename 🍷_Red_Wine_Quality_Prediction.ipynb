```python
### As the data consists of values which differ a lot in magnitude they need to be brought to same scale its done using standard scaler
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
new_data=scaler.fit_transform(df.drop(labels=['quality'],axis=1))
```

```python
df.columns
```

```python
columns=['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol']
```

```python
new_df=pd.DataFrame(data=new_data,columns=columns)
```

```python
new_df
```

```python
## AS There are no missing values so we dont need to do data imputation
```

```python
x=new_df
y=df['quality']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3, random_state=42)
```

```python
from sklearn.svm import SVC
model=SVC()
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,model.predict(x_test))
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_pred, y_test))
```

```python
### AS observed the accuracy of the moedel is quit low we need to implement the grid search approach to optimize the parameters to give the best accuracy/
```

# **Implementating Grid Serch CV**

```python
from sklearn.model_selection import GridSearchCV
```

```python
Param_grid={
    'C':[0.1,1,10,50,100,500],
    'gamma':[1,0.1,0.01,0.001]
}
```

```python
grid=GridSearchCV(SVC(),Param_grid,verbose=3,n_jobs=-1)
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
model_new=SVC(C=10,gamma=1)
model_new.fit(x_train,y_train)
```

```python
y_pred=model_new.predict(x_test)
```

```python
y_pred
```

```python
accuracy_score(y_test,model_new.predict(x_test))
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_pred, y_test))
```
